## Refine email fetching (imap)

### Subtask:
Modify the existing code to fetch a larger set of emails for analysis and categorization. Ensure proper handling of email parts and encodings.


**Reasoning**:
Modify the existing code to fetch all emails and extract the subject and body into a list of dictionaries for later use.



```python
import ssl
import email

# Initialize a list to store email data
email_data = []

try:
    # Force SSL context (fixes certificate issues)
    context = ssl.create_default_context()

    # Connect with timeout
    server = IMAPClient(
        'imap.gmail.com',
        ssl=True,
        ssl_context=context,
        timeout=10  # Add timeout
    )
    print("Connected to server.")

    # Login
    server.login('arj831758@gmail.com', 'lntw kycc retq yuit')
    print("Login successful.")

    # Select folder
    server.select_folder('INBOX')
    print("INBOX selected.")

    # Search for all emails
    messages = server.search(['ALL'])
    print(f"Found {len(messages)} emails.")

    if not messages:
        print("No emails found with the specified criteria.")
    else:
        for msg_id, data in server.fetch(messages, ['RFC822']).items():
            raw_email = data[b'RFC822']
            msg = email.message_from_bytes(raw_email) # Parse the raw email data

            subject = msg.get('Subject')
            body = ""

            if msg.is_multipart():
                for part in msg.walk():
                    ctype = part.get_content_type()
                    cdispo = part.get('Content-Disposition')

                    # Look for plain text or HTML parts, and ignore attachments
                    if ctype == 'text/plain' and cdispo is None:
                        try:
                            body = part.get_payload(decode=True).decode('utf-8', errors='ignore')
                            break # Prioritize plain text
                        except Exception as e:
                            print(f"Error decoding plain text part for email ID {msg_id}: {e}")
                            body = "Error decoding body"
                    elif ctype == 'text/html' and cdispo is None:
                        try:
                            html_body = part.get_payload(decode=True).decode('utf-8', errors='ignore')
                            # Store HTML body if no plain text is found
                            if not body:
                                body = html_body
                        except Exception as e:
                            print(f"Error decoding HTML part for email ID {msg_id}: {e}")
                            if not body:
                                body = "Error decoding body"

            else:
                # Not a multipart message, likely a single part
                try:
                    body = msg.get_payload(decode=True).decode('utf-8', errors='ignore')
                except Exception as e:
                    print(f"Error decoding single part email ID {msg_id}: {e}")
                    body = "Error decoding body"

            email_data.append({'id': msg_id, 'subject': subject, 'body': body})

except Exception as e:
    print(f"Error: {e}")
finally:
    if 'server' in locals() and server.has_capability('LOGOUT'):
        server.logout()

# You can now process the email_data list for categorization
# For demonstration, print the first 5 entries
print("\n--- Extracted Email Data (first 5) ---")
for i, email_info in enumerate(email_data[:5]):
    print(f"ID: {email_info['id']}")
    print(f"Subject: {email_info['subject']}")
    print(f"Body (first 200 chars): {email_info['body'][:200]}...")
    print("-" * 20)
```

## Prepare data for ai/ml

### Subtask:
Extract relevant information from the fetched emails (subject, body text) and structure it for use with NLP and classification models.


**Reasoning**:
Import pandas, convert the list of dictionaries to a DataFrame, handle missing values, combine subject and body into a single text column, clean the combined text, and display the head of the DataFrame.



```python
import pandas as pd
import re
import string

# Convert the list of email data dictionaries into a pandas DataFrame
df_emails = pd.DataFrame(email_data)

# Handle potential missing values by filling with empty strings
df_emails['subject'] = df_emails['subject'].fillna('')
df_emails['body'] = df_emails['body'].fillna('')

# Combine subject and body into a single text column
df_emails['combined_text'] = df_emails['subject'] + ' ' + df_emails['body']

# Clean and preprocess the combined text data
def clean_text(text):
    text = text.lower() # Convert to lowercase
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text) # Remove punctuation
    text = re.sub(r'\s+', ' ', text).strip() # Remove extra whitespace
    return text

df_emails['cleaned_text'] = df_emails['combined_text'].apply(clean_text)

# Display the first few rows and check data types
display(df_emails.head())
display(df_emails.info())
```

## Choose and implement categorization method

### Subtask:
Decide on the criteria for categorizing emails (keywords, sentiment, or a trained model) and implement the chosen method using appropriate libraries (NLTK, spaCy, scikit-learn, etc.).


**Reasoning**:
I will implement keyword matching to categorize emails as "Success", "Rejected", or "Neutral/Other" based on the cleaned text, as this is a straightforward approach given the task doesn't specify a need for nuanced sentiment or complex patterns requiring a trained model.



```python
# Define keyword lists for categorization
success_keywords = ['successful', 'accepted', 'offer', 'congratulations', 'welcome', 'approved', 'hired']
rejected_keywords = ['rejected', 'unsuccessful', 'regret', 'unfortunately', 'not selected', 'moved forward', 'other candidates']

# Function to categorize emails based on keywords
def categorize_email(text):
    text_lower = text.lower()
    if any(keyword in text_lower for keyword in success_keywords):
        return 'Success'
    elif any(keyword in text_lower for keyword in rejected_keywords):
        return 'Rejected'
    else:
        return 'Neutral/Other'

# Apply the categorization function to the cleaned text
df_emails['category'] = df_emails['cleaned_text'].apply(categorize_email)

# Display the value counts for the new category column
display(df_emails['category'].value_counts())

# Display the first few rows with the new category column
display(df_emails[['subject', 'cleaned_text', 'category']].head())
```

```python
import ssl
from imapclient import IMAPClient
import time

# Re-connect and expunge to finalize deletions

try:
    # Force SSL context
    context = ssl.create_default_context()

    # Connect with a potentially longer timeout for robustness
    server = IMAPClient(
        'imap.gmail.com',
        ssl=True,
        ssl_context=context,
        timeout=30  # Increased timeout
    )
    print("Re-connected to server for expunge.")

    # Login
    server.login('arj831758@gmail.com', 'lntw kycc retq yuit')
    print("Login successful.")

    # Select inbox folder
    server.select_folder('INBOX')
    print("INBOX selected for expunge.")

    # Perform expunge to permanently delete marked messages
    # Check if there are messages marked for deletion before expunging
    # Use folder_status to get status information for the selected folder
    status_data = server.folder_status('INBOX', ['DELETED'])
    # The response is a dictionary where the key is the folder name (bytes)
    # and the value is a dictionary of status items.
    deleted_count = status_data[b'INBOX'][b'DELETED']

    print(f"Number of messages marked for deletion in INBOX: {deleted_count}")

    if deleted_count > 0:
        server.expunge()
        print(f"Expunged {deleted_count} messages from INBOX.")
    else:
        print("No messages marked for deletion to expunge.")


except Exception as e:
    print(f"An error occurred during re-connection and expunge: {e}")
finally:
    if 'server' in locals() and server.has_capability('LOGOUT'):
        try:
            server.logout()
            print("Logged out from server after expunge attempt.")
        except Exception as e:
            print(f"Error during final logout: {e}")
    elif 'server' in locals():
         # Fallback for servers that don't support or fail on LOGOUT
         # Note: IMAPClient's logout is generally preferred and handles closing
         print("Server connection ended after expunge attempt.")

# Assuming the expunge was successful, the task of moving and deleting is complete.
# No dataframes were used in this subtask.
```
